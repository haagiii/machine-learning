```python
import numpy as np
import pydotplus
from sklearn.feature_extraction.text import TfidfVectorizer
from IPython import display
from sklearn import tree
from io import StringIO
from operator import itemgetter
%matplotlib inline
%load_ext autoreload
%autoreload 2



input_text=np.array([
    '物理 数学 国語　　', #　ドキュメント1
    'パン ライス 　チョコ', #ドキュメント2
    'ビル アパート　マンション　', #ドキュメント3
    '北海道 福岡　神奈川' #ドキュメント4
])

#ドキュメント1は1
#ドキュメント2は2
#ドキュメント3は3
#ドキュメント4は4

target_data = np.array([
    1,2,3,4
    ])

#文書をtf-idfのベクトルへ変換
vectorizer=TfidfVectorizer()

input_data=vectorizer.fit_transform(input_text)

np.set_printoptions(precision=2) # 有効表示桁数を2桁へ変更
print(input_data.toarray())

#決定木モデルを生成
model = tree.DecisionTreeClassifier(max_depth=5)
model = model.fit(input_data, target_data)

model.predict(input_data)

feature_names = list(map(itemgetter(0),
sorted(vectorizer.vocabulary_.items(), key=itemgetter(1))))

data = StringIO()
tree.export_graphviz(model, out_file=data, feature_names=feature_names)
graph = pydotplus.graph_from_dot_data(data.getvalue())

display.display(display.Image(graph.create_png()))
```

```python
import numpy as np
import pydotplus
from sklearn.feature_extraction.text import TfidfVectorizer
from IPython import display
from sklearn import tree
from sklearn.datasets import load_digits # https://pythondatascience.plavox.info/scikit-learn/scikit-learn%E3%81%AB%E4%BB%98%E5%B1%9E%E3%81%97%E3%81%A6%E3%81%84%E3%82%8B%E3%83%87%E3%83%BC%E3%82%BF%E3%82%BB%E3%83%83%E3%83%88
from sklearn.model_selection import train_test_split # http://www.madopro.net/entry/skleran_train_test_split
from io import StringIO
from operator import itemgetter
%matplotlib inline
%load_ext autoreload
%autoreload 2


digits = load_digits(n_class=10)

# 訓練データとテストデータを8:2で分割
input_data, test_data, target_data, T_test = train_test_split(
    digits.data, digits.target, test_size=0.2)
print(input_data)
print(len(input_data))
print(target_data)
print(len(target_data))


np.set_printoptions(precision=2) # 有効表示桁数を2桁へ変更

#決定木モデルを生成
model = tree.DecisionTreeClassifier(max_depth=5)
model = model.fit(input_data, target_data)
model.predict(test_data)

# こちらはアンサンブル学習用のランダム決定木らしい。
model2 = tree.ExtraTreeClassifier()  (max_depth=5)
model2 = model2.fit(input_data, target_data)
model2.predict(test_data)

# 表示
data = StringIO()
tree.export_graphviz(model, out_file=data)
graph = pydotplus.graph_from_dot_data(data.getvalue())
display.display(display.Image(graph.create_png()))

data2 = StringIO()
tree.export_graphviz(model2, out_file=data2)
graph = pydotplus.graph_from_dot_data(data2.getvalue())
display.display(display.Image(graph.create_png()))
```
